
# Download texts

```python
import requests
import os
from bs4 import BeautifulSoup
import json


max_id = 50000


for i in range(1,(max_id+1)):
    url = f'https://bnm-i.huygens.knaw.nl/api/domain/ebnmteksten/TEXT{ str(i).rjust(12, "0" )  }'
    print(url)

    response = requests.get(url)
    TEXT_id = os.path.basename(url)
    
    if response.status_code == 200:
        response.encoding = 'utf-8'
        html_page = response.text
        soup = BeautifulSoup( html_page,"lxml")

        p = soup.find('p')
        
        path = os.path.join( 'BNM_texts' , f'{TEXT_id}.txt' ) 

        out = open( path , 'w' , encoding = 'utf-8' )
        out.write(p.text)
        out.close()

```

```python
import requests
import os
from bs4 import BeautifulSoup
import json

compound = ['annotatie_incipit',
            'annotatie_overig',
           'sleutelw_incipit',
            'titel_hs_inc_form' ,
            'annotatie_inhoud' ,
            '@variationRefs' , 
            '@relations']

ignore = ['^created' , 
          '^modified',
          '^rev' , 
         '^deleted' , 
          '@relationCount' , 
          '@properties' ]

fields = ['text_part_of',
          'text_has_categorie',
          'text_has_doc',
          'tekstIsPartOfTekstdragerWithRegiocode',
          'tekstIsPartOfTekstdragerWithLocalisering',
          'text_has_trefwoord']


nr_id = 30000

'TEXT000000008331'

for i in range(8330,(nr_id+1)):
    url = f'https://bnm-i.huygens.knaw.nl/api/domain/ebnmteksten/TEXT{ str(i).rjust(12, "0" )  }'


    response = requests.get(url)
    bnm_id = os.path.basename(url)
    if bnm_id == 'TEXT000000008331':
        print('YES!!')
    
    if i % 50 == 0:
        print(i , end = ';')
    #print(bnm_id)

    if response:
        print(url)
        response.encoding = 'utf-8'
        html_page = response.text
        soup = BeautifulSoup( html_page,"lxml")

        p = soup.find('p')

        json_data = json.loads(p.text)

        out = open( f'BNM_data/{bnm_id}.txt' , 'w' , encoding = 'utf-8' )

        for item in json_data:
            if (not item in compound) and (not item in ignore): 
                out.write( f'{item}\t{json_data[item]}\n' )

        for i in range(0,5):
            field = compound[i]
            for i,a in enumerate( json_data[field]):
                out.write( f'{field}_{i}\t{a}\n' )

        relations_dict = json_data['@relations']

        for f in fields: 
            out.write( f'#### relations -- {f} ####\n' )

            for item in relations_dict[f]:
                for key in item.keys():
                    out.write( f'relations_{f}_{key}\t{item[key]}\n')


        out.close()


        
```
